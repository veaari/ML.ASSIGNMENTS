```python
import pandas as pd

# Replace 'your_dataset.csv' with the path to your dataset
df = pd.read_csv('Pharma_data.csv')
```

```python
# Example data cleaning steps
df.dropna(inplace=True)  # Remove rows with missing values
# Convert data types if necessary
df['Effectiveness'] = df['Effectiveness'].astype(float)
df['Side_Effects'] = df['Side_Effects'].astype(float)
df['Marketing_Spend'] = df['Marketing_Spend'].astype(float)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
sns.barplot(x='Region', y='Effectiveness', hue='Product_ID', data=df)
plt.title('Average Effectiveness by Region and Product')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.violinplot(x='Product_ID', y='Effectiveness', data=df, inner='quartile')
plt.title('Distribution of Effectiveness by Product')
plt.show()

plt.figure(figsize=(12, 6))
sns.violinplot(x='Product_ID', y='Side_Effects', data=df, inner='quartile')
plt.title('Distribution of Side Effects by Product')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.violinplot(x='Product_ID', y='Effectiveness', data=df, inner='quartile')
plt.title('Distribution of Effectiveness by Product')
plt.show()

plt.figure(figsize=(12, 6))
sns.violinplot(x='Product_ID', y='Side_Effects', data=df, inner='quartile')
plt.title('Distribution of Side Effects by Product')
plt.show()

```

```python
sns.pairplot(df, vars=['Effectiveness', 'Side_Effects', 'Marketing_Spend'], hue='Product_ID')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x='Trial_Period', y='Effectiveness', data=df)
plt.title('Effectiveness by Trial Period')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.regplot(x='Marketing_Spend', y='Effectiveness', data=df)
plt.title('Marketing Spend vs. Effectiveness')
plt.show()
```
